```python
def show_nan(data):
    """Функция берет на вход датафрейм, считает сумму пропусков в нем, затем считает и
    выводит долю пропусков и общее их количество в столбцах, сортируя по убыванию"""
    nan = (data.isna() | data.isnull()).sum()
    result = (pd.concat([nan / data.shape[0], nan], axis=1,  keys=['ratio', 'nan_count'])
           .sort_values('ratio', ascending=False))
    return result
```

### 1.2. Проверю, что эффективность обогащения рассчитана правильно. Вычислю её на обучающей выборке для признака `rougher.output.recovery`. Найду MAE между моими расчётами и значением признака. Опишу выводы.

***Эффективность обогащения*** рассчитывается по формуле:

$$Recovery = \frac{C * (F - T)}{F * (C - T)} * 100 % $$

где:

$1. \:C\: — \:доля \:золота \:в \:концентрате \:после \:флотации/очистки$

$2. \:F\: — \:доля \:золота \:в \:сырье/концентрате \:до \:флотации/очистки$

$3. \:T\: — \:доля \:золота \:в \:отвальных \:хвостах \:после \:флотации/очистки$

Рассчитаем эффективность обогащения чернового концентрата `rougher.output.recovery` из обучающей выборки. Предварительно удалим пропуски из столбцов со значением NaN:

```python
train_df.dropna(subset=['rougher.output.concentrate_au', 'rougher.input.feed_au',
                                   'rougher.output.tail_au','rougher.output.recovery'], inplace=True)
```

```python
# Инициализирую переменные

C = train_df['rougher.output.concentrate_au']
F = train_df['rougher.input.feed_au']
T = train_df['rougher.output.tail_au']
```

```python
# Рассчитываю эффективность обогаения на обучающей выборке

train_df['rougher.output.recovery_calc'] = C * (F - T) / (F * (C - T)) * 100
```

```python
# Рассчитаю MAE(среднюю абсолютную ошибку) между моими расчетами и расчетами признака, но
# т.к. в наших данных есть пропуски и при их удалении пропусков получаются разные размеры наших датафреймов, 
# то тут готовая формула из библиотеки sklearn не подойдёт. Напишем свою формулу вручную:

def mae(target_recovery, calculated_recovery):
    return (abs(target_recovery - calculated_recovery).sum() / len(target_recovery))

target_recovery = train_df['rougher.output.recovery']
calculated_recovery = train_df['rougher.output.recovery_calc']

print('Средняя абсолютная ошибка (MAE) =', mae(target_recovery, calculated_recovery))
```

**Вывод:**
    
    Как мы видим средняя абсолютная ошибка достаточно мала (чем меньше значение MAE, тем лучше), считаем что recovery в train рассчитан верно.

```python
def get_collinear_features(df, threshold):
    '''
    Описание:
        Функция обходит матрицу корреляций, сранивая попарно признаки в датафрейме.
        Если корреляция между признаками выше установленного порога, то такие данные
        мультиколлинеарные.
        
    Входные данные:
        df: датафрейм с признаками
        threshold: Порог корреляции, когда признаки считаются сильно связанными между собой
    
    Выход: 
        массив, который содержит пары мультиколлинеарных признаков
    '''
    
    # Удаление целевой переменной из датафрейма   
    x = df.drop('rougher.output.recovery', axis=1)
    
    # Расчет матрицы корреляций
    corr_matrix = x.corr()    
    iters = range(len(corr_matrix.columns))   
    drop_cols = []

    # Обход матрицы, попарно сравнивая признаки между собой
    for i in iters:        
        for j in range(i+1):            
            item = corr_matrix.iloc[j:(j+1), (i+1):(i+2)]          
            col = item.columns            
            row = item.index            
            val = abs(item.values)           
            
            # Если корреляция превышает порог
            if val >= threshold:                
                drop_cols.append((col.values[0], row.values[0], round(val[0][0], 2))) 
      
    return drop_cols
```

```python
deleted_cols = set() # удаленные столбцы из пар мультиколлинеарных признаков

for pair in multi_list:
    
    not_deleted = True
    if pair[0] in columns_not_in_test:
        delete_col_from_pair(train_df, pair[0])
        delete_col_from_pair(test_df, pair[0])
        delete_col_from_pair(full_df, pair[0])
        not_deleted = False
        deleted_cols.add(pair[0])
        
    if pair[1] in columns_not_in_test and not_deleted:
        delete_col_from_pair(train_df, pair[1])
        delete_col_from_pair(test_df, pair[1])
        elete_col_from_pair(full_df, pair[1])
        not_deleted = False
        deleted_cols.add(pair[1])
        
    if (pair[0] not in columns_not_in_test) and (pair[1] not in columns_not_in_test) and not_deleted:
        delete_col_from_pair(train_df, pair[0])
        delete_col_from_pair(test_df, pair[0])
        delete_col_from_pair(full_df, pair[0])
        not_deleted = False
        deleted_cols.add(pair[0])
```

### 3.1. Напишу функцию для вычисления итоговой sMAPE.

Метрика качества, необходимая для оценки эффективности процесса очистки золотосодержащей руды — `sMAPE` (симметричное среднее абсолютное процентное отклонение).  

Метрика sMAPE вычисляется так:


$$sMAPE = \frac{1}{N} \sum_{i=1}^{N}\frac{\left | y_{i} -\
\hat{y}_{i}\right |}{\left ( \left | y_{i} \right | +\left | \hat{y}_{i} \right |\right )/2} *100\%$$  



 - Значение `целевого признака` для объекта с порядковым номером i в выборке, на которой измеряется качество: $$y_{i}$$   

 - Значение `предсказания` для объекта с порядковым номером i, например, в тестовой выборке: $$\hat{y}_{i}$$  

 - Количество `объектов` в выборке: $$N$$  

 - `Суммирование по всем объектам выборки` (i меняется от 1 до N). $$\sum_{i=1}^{N}$$

```python
def smape_calc(target, prediction):
    """Функция вычисления метрики smape"""
    diff = abs(target - prediction)
    add = (abs(target) + abs(prediction))/2
    smape_stage = 1/len(target)*sum(diff/add)*100
    
    return smape_stage
```

```python
#def final_smape(smape_r, smape_f):
#    """Функция вычисления итоговой метрики smape"""
#    final_smape = 0.25*smape_r + 0.75*smape_f
#    return final_smape
```

```python
def final_smape(target, predictions):
    """Функция вычисления итоговой метрики smape"""
    smape_r = smape_calc(target[0], predictions[0])
    smape_f = smape_calc(target[1], predictions[1])
    final_smape = 0.25*smape_r + 0.75*smape_f
    return final_smape
```
